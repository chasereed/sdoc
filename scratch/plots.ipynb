```python
import base64
import io
import matplotlib.pyplot as plt


plt.plot([0, 1], [0, 1])

fs = io.BytesIO()
plt.savefig(fs, format='png')
image = fs.getvalue()
image_b64 = base64.b64encode(image).decode('utf-8')
print(image)

# Display in Jupyter using the HTML image tag
from IPython.display import HTML
HTML(f'<img src="data:image/png;base64,{image_b64}">')
```

```python
# create new axes
plt.close()
fig, ax = plt.subplots()
ax.plot([0, 1], [0, 1])

def plot_to_html(figure=None):
    if figure is None:
        figure = plt.gcf()
    img = io.BytesIO()
    figure.savefig(img, format='png')
    img.seek(0)
    return f'<img src="data:image/png;base64,{base64.b64encode(img.read()).decode()}">'


display(HTML(plot_to_html()))

```
